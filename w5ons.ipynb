```python
# // ONS DOWNLOADER
import requests
import json
from google.colab import files

url_base = "https://api.ons.gov.uk/timeseries/{}/dataset/{}/data"
file_base = "data_ONS-{}.json"
SeriesCode = ['BOKG', 'BOKH', 'BOKI', 'IKBB', 'IKBD', 'IKBC', 'IKBH', 'IKBI', 'IKBJ']
Dataset = ['MRET', 'PN2', 'QNA', 'QNA', 'DIOP', 'UKEA', 'PN2', 'UKEA', 'UKEA']

for (i, j) in zip(SeriesCode, Dataset):
   URL = url_base.format(i, j) #Update URL 
   data = requests.get(URL).json() # Get Data
   fileName = file_base.format(i) #Update Filename
   with open(fileName, 'w', encoding='utf-8') as f:
     json.dump(data, f, ensure_ascii=False, indent=4) #Open Data into Colab
   files.download(fileName) #Download 
```
